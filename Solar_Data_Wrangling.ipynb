```python
# Impute missing values using mean imputation for numeric columns
from sklearn.impute import SimpleImputer
numeric_columns = ['Humidity', 'AmbientTemp', 'Wind_Speed', 'Visibility', 'Pressure', 'Cloud.Ceiling']
imputer = SimpleImputer(strategy='mean')
df[numeric_columns] = imputer.fit_transform(df[numeric_columns])
```

```python
filter = df[(df['Season'] == 'Summer') & (df['Hour'] == 12)]
filter
```

```python
df['Year'] = df['YRMODAHRMI'] // 100000000
df['Month'] = (df['YRMODAHRMI'] // 1000000) % 100
df['Day'] = (df['YRMODAHRMI'] // 10000) % 100
df['Hour'] = (df['YRMODAHRMI'] // 100) % 100
df['Minute'] = df['YRMODAHRMI'] % 100

df
```
